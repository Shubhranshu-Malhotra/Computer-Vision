## Blending/Overlaying images of "different sizes"

### Overlay small image on top of larger image [No Blending]
**Using numpy reassignment**

```python
img_1 = cv2.imread('DATA/00-puppy.jpg')
img_1 = img_1[:,:,::-1]
plt.imshow(img_1)
```

```python
img_2 = cv2.imread('DATA/watermark_no_copy.png')
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
plt.imshow(img_2)
```

```python
large_image = img_1.copy()
small_image = cv2.resize(img_2, (800,600))
large_image.shape, small_image.shape
```

```python
x_offset = 0
y_offset = 0
x_end = x_offset + small_image.shape[1]
y_end = y_offset + small_image.shape[0]
```

```python
# x_end
```

```python
plt.imshow(large_image)
```

```python
large_image[10:11,10:12]
```

```python
# y_partition(height) is first as img[rows,cols]: rows = height, cols = width
large_image[y_offset:y_end,x_offset:x_end] = small_image
```
